## CPU vs GPU (CUDA)

```python
torch.cuda.is_available()
```

# PyTorch computational graph
Important property of PyTorch tensor operations - they create a **dynamic computational graph**.  
The tensor **c** knows that how it was created from **a** and **b**.  
Thanks to this, derivatives (gradients) can be computed for tensors.  
Example:  
<img src="tree-eval.png" alt="Drawing" style="width: 600px;"/>  
*Source: http://colah.github.io/posts/2015-08-Backprop/ - good blog post to understand backpropagation*

### Compute for a=2, b=1

```python
# Create a and b
# Compute c, d, e
a = torch.tensor(2.0, requires_grad=True) # requires_grad is true so u can see the grad function of the calculations 
b = torch.tensor(1.0, requires_grad=True)
c = a + b
d = b + 1
e = c * d
print(a, b, c, d, e)
#this is the tree from above
```

```python
# grad_fn stores information about computation
print(c.grad_fn)
print(e.grad_fn)
```

### Now we can computer derivative of $e$ w.r.t. $a, b, c, d$
<img src="tree-eval-derivs.png" alt="Drawing" style="width: 600px;"/>  

\begin{equation} 
e = c * d
\end{equation}
\begin{equation} 
\frac {\partial e} {\partial c} = d = 2
\end{equation}
\begin{equation} 
\frac {\partial e} {\partial d} = c = 3
\end{equation}
\begin{equation} 
\frac {\partial e} {\partial a} = \frac {\partial e} {\partial c} \frac {\partial c} {\partial a} = 2
\end{equation}
\begin{equation} 
\frac {\partial e} {\partial b} = \frac {\partial e} {\partial c} \frac {\partial c} {\partial b} + \frac {\partial e} {\partial d} \frac {\partial d} {\partial b} = 2 * 1 + 3 * 1 = 5
\end{equation}

Compute gradients with `torch.autograd.grad` https://pytorch.org/docs/stable/autograd.html

```python
# Compute gradients with torch.autograd.grad https://pytorch.org/docs/stable/autograd.html
```

```python
grad_a, grad_b, grad_c, grad_d = torch.autograd.grad(e, [a, b, c, d]) # get the gradient of e with respect to abcd
```

```python
print(grad_a, grad_b, grad_c, grad_d)
```

```python
# Another way for leaf tensors with `backward` method
```

```python
# Create a and b
# Compute c, d, e
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)
c = a + b
d = b + 1
e = c * d
print(a, b, c, d, e)
```

```python
e.backward() #lets say e is our loss (which is the final node of our computation) and we want to know its gradient, we can just do e.backward instead of torch.autograd.grad(..) and it will automatically compute all the gradients of all the weights of the nodes that serve in the calc of this e
```

```python
print(a.grad, b.grad, c.grad, d.grad) #but it doesnt show you it intermediate nodes, we only see the leaf nodes. We only need the gradients for our weights, which are the leaf nodes so it is fine.
```
